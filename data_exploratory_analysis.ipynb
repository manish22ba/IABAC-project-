## Data Processing/ Data Munging

```python
#convert the Categorical Variable into Numerical by using LabelEncoder
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
lab_enc_Dataset=df.apply(encode.fit_transform)
lab_enc_Dataset.head()
```

```python
# Select features based on k highest scores.
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
lab_enc_Dataset.shape
```

```python
#independent_features
X= lab_enc_Dataset.iloc[:,:-1]

#dependent_features
Y = lab_enc_Dataset['PerformanceRating']
```

```python
### Applying SelectKBest Algorithm
ordered_rank_features=SelectKBest(score_func=chi2,k=20)
ordered_feature=ordered_rank_features.fit(X,Y)
```

```python
df_VALUES=pd.DataFrame(ordered_feature.scores_,columns=["Score"])
columns_name=pd.DataFrame(X.columns)
```

```python
features_rank=pd.concat([columns_name,df_VALUES],axis=1)
features_rank.columns=['Features','k_values']
features_rank
```

```python
features_rank.nlargest(5,'k_values')
```

### Insigths from the above SelectKBest Algorithm 
- The above mentioned features are top 5 Features with '**k highest values**'

## Feature Importance
- Now by Feature importance we will see the values for each feature of our data,the higher the score more relevant it is.

```python
from sklearn.ensemble import ExtraTreesClassifier
model=ExtraTreesClassifier()
model.fit(X,Y)
```

```python
print(model.feature_importances_)
```

```python
ranked_features=pd.Series(model.feature_importances_,index=X.columns)
ranked_features.nlargest(10).plot(kind='barh', title = 'Feature importanceâ€™s coeffecients plot')
plt.show()
```

### Insigths from the above Feature Importance
We can observe from the TOP:5 features which affect the Performance Rating of an Employee are below :
1.  EmpLastSalaryHikePercent
2.  YearsSinceLastPromotion
3.	EmpEnvironmentSatisfaction
4.	ExperienceYearsAtThisCompany
5.	ExperienceYearsInCurrentRole
